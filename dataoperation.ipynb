```python
colors=['red','green','blue']
species=['Iris-setosa','Iris-versicolor','Iris-virginica']
#defining species and colors to plot scatter graph
```

```python
#to plot it for sepal lenght vs sepal width
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'], c=colors[i], label=species[i])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
#to plot it for petal lenght vs petal width
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['PetalLengthCm'], x['PetalWidthCm'], c=colors[i], label=species[i])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
#to plot it for sepal lenght vs petal lenght
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLengthCm'], x['PetalLengthCm'], c=colors[i], label=species[i])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

```python
#to plot it for sepal width vs petal width
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalWidthCm'], x['PetalWidthCm'], c=colors[i], label=species[i])

plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
#now model training
#training data - 70%
#testing data - 30%
X=df.drop(columns=['Species'])
Y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train) #this is model training
```

```python
print("accuracy score - ",model.score(x_test,y_test)*100)
```

```python
#importing knn as other model
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(x_train,y_train)
```

```python
print("accuracy score - ",model2.score(x_test,y_test)*100)
```

```python
#importing high level module
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
```

```python
print("accuracy score - ",model3.score(x_test,y_test)*100)
```
